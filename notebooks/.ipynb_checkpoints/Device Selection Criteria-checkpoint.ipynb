```python
from pathlib import Path

import pandas as pd
import altair as alt

import ideafast_deviceselection as ifds

source = Path(ifds.__file__).parent.parent.absolute() / 'local/DeviceSelectionScoring.xlsx'


df = pd.read_excel(
    f"{source}",
    index_col=0, header=[0,1,2], nrows=63,
    sheet_name='combined scores')
df = df.drop(df.columns[:4],axis=1)

# extend empty headers to use the multiIndex
a = df.columns.get_level_values(0)
b = df.columns.get_level_values(1)
c = df.columns.get_level_values(2)
df.columns = [a.to_series().mask(lambda x: x.str.startswith('Unnamed')).ffill(), b, c]

df = df.reset_index()

df
```

```python
df.pivot(columns='index')
```

```python
long = pd.melt(df.reset_index(), id_vars=['index'], var_name=['device','stream','type'])

long.pivot(index=['index','device','stream'], columns='type')
```

```python
vtp = long[long["device"] == "Vital Patch"]

# pivoted = pd.pivot(vtp, index=['index', 'device','stream'], columns='type')

pivoted = vtp.unstack()

pivoted
```

```python
alt.Chart(pivoted).mark_point().encode(
    x='stream',
    y='count(value)'
)
```
