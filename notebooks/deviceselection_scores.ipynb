```python
import pandas as pd
import altair as alt

df = pd.read_excel(
    'DeviceSelectionScoring.xlsx',
    index_col=0, header=[0,1,2], nrows=63,
    sheet_name='combined scores')
df = df.drop(df.columns[:4],axis=1)

# extend empty headers to use the multiIndex
a = df.columns.get_level_values(0)
b = df.columns.get_level_values(1)
c = df.columns.get_level_values(2)
df.columns = [a.to_series().mask(lambda x: x.str.startswith('Unnamed')).ffill(), b, c]
df.index.names = ['Criteria']
df.columns.names = ['device','stream','value']

df
```

```python
vtp = df.loc[slice(None),('Vital Patch',slice(None),slice(None))]

vtp
```

```python
alt.Chart(vtp).mark_point().encode(
    alt.X(alt.repeat(), type='nominal')
).repeat(['Vital Patch', 'experts (e)', 'score']
```
